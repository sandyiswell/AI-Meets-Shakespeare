```python
# Importint the necessary package.
from bert_score import score
```

```python
# 'candidates' : The output of the LLM.
candidates = [
  """
  CASSIUS.
But wait! I ask you. Did Caesar faint?

CASCA.
He collapsed in the marketplace, foaming at the mouth and unable to speak.

BRUTUS.
It’s very likely; he’s suffering from epilepsy.

CASSIUS.
No, Caesar doesn’t have it; but you, me, and Casca do

  """  # Reference 1
]

# 'reference' : The human translated portion pertaining to the candidate output. Taken from LitChart.
references = [
   """
   CASSIUS
    Stop for a moment, please. What, did Caesar faint?

    CASCA
    He fell down in the marketplace and foamed at the mouth and couldn't speak.

    BRUTUS
    That’s very likely. He has epilepsy, the falling sickness.

    CASSIUS
    No, Caesar doesn’t have it. But you and I, and honest Casca, we have the falling sickness.

   """

]


# Compute BERTScore
P, R, F1 = score(candidates, references, lang="en", verbose=True)

# Display the results
for i, (precision, recall, f1) in enumerate(zip(P, R, F1)):
    print(f"Sentence {i+1}:")
    print(f"  Precision: {precision:.4f}")
    print(f"  Recall: {recall:.4f}")
    print(f"  F1 Score: {f1:.4f}")

```
