Postprocesser

```python
import os

# Folder containing the text files
folder_path = "./preprocessed_data"
output_file = "V110_output_combined.txt"
delimiter = "**********"
words_to_check = {"Please", "Shakespearean", "Note", "provide"}

# Function to check if any of the last two lines contain the specified words
def should_omit_line(line):
    return any(word in line for word in words_to_check)

# Open the output file to write the filtered content
with open(output_file, "w") as outfile:
    for filename in os.listdir(folder_path):
        if filename.startswith("V110_Julius"):  # Process only text files
          print(filename)
          file_path = os.path.join(folder_path, filename)
          with open(file_path, "r") as infile:
              content = infile.read()

              # Split content by the delimiter and get text after it
              parts = content.split(delimiter)
              if len(parts) > 1:
                  text_after_delimiter = parts[1].strip()

                  # Split into lines and check the last two lines
                  lines = text_after_delimiter.split("\n")
                  if len(lines) > 2:
                      if should_omit_line(lines[-1]):
                          lines.pop(-1)
                      if should_omit_line(lines[-1]):
                          lines.pop(-1)

                  # Write the remaining lines to the output file
                  outfile.write("\n".join(lines) + "\n")
```
